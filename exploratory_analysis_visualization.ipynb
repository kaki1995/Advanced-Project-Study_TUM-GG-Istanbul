#### Clean and standardize tool names in q10_tool_1 to q10_tool_4 in df_gg_surveys

```python
# List unique values and their counts for q10_tool_1 to q10_tool_4 in df_gg_surveys
tools = []
for col in ["q10_tool_1", "q10_tool_2", "q10_tool_3", "q10_tool_4"]:
    if col in df_gg_surveys.columns:
        tools.extend(df_gg_surveys[col].dropna().astype(str).tolist())
tool_counts = Counter(tools)
print("Unique tool values and their counts across q10_tool_1 to q10_tool_4:")
for tool, count in tool_counts.items():
    print(f"{tool}: {count}")
```

```python
# Clean and standardize tool names in q10_tool_1 to q10_tool_4 in df_gg_surveys, with refined logic for 'Google target' in q10_tool_2
def clean_and_split_tools(row):
    replacements = {
        'Google target': None,  # handled separately below
        'directly on target shop website': 'Official websites',
        'Ecommerce site (Shopee)': 'Shopee',
        'official website': 'Official websites',
        'Chatgpt': 'ChatGPT'
    }
    # Clean each tool column
    for col in ["q10_tool_1", "q10_tool_2", "q10_tool_3", "q10_tool_4"]:
        val = row.get(col, None)
        if pd.isna(val):
            continue
        val = str(val)
        if col == "q10_tool_2" and val == "Google target":
            row["q10_tool_2"] = "Google"
            # Only set q10_tool_3 if it is empty or NaN
            if pd.isna(row.get("q10_tool_3", None)) or row["q10_tool_3"] == '':
                row["q10_tool_3"] = "Target"
            else:
                row["q10_tool_3"] = str(row["q10_tool_3"]) + ", Target"
            continue
        # Apply replacements for other values
        if val in replacements and replacements[val] is not None:
            row[col] = replacements[val]
        else:
            row[col] = val
    return row

tool_cols = ["q10_tool_1", "q10_tool_2", "q10_tool_3", "q10_tool_4"]
if all(col in df_gg_surveys.columns for col in tool_cols):
    df_gg_surveys[tool_cols] = df_gg_surveys.apply(clean_and_split_tools, axis=1)[tool_cols]

# Preview the affected columns to confirm
df_gg_surveys[tool_cols].head(10)
```

### 3.2 What kind of device participants used for the experiment?

```python
# 100% stacked horizontal bar chart: Device Type (GG upper, Chat lower) with internal percentage labels and ARD/YouGov style
def get_device_percentages(df):
    counts = df['device_type'].value_counts()
    total = counts.sum()
    mobile = counts.get('mobile', 0)
    desktop = counts.get('desktop', 0)
    return [mobile / total * 100 if total else 0, desktop / total * 100 if total else 0]

gg_pct = get_device_percentages(df_gg_surveys)
chat_pct = get_device_percentages(df_chat_surveys)

# prepare plotting values (order: Mobile, Desktop)
mobile_color = '#f1c40f'  # gold-like for Mobile
desktop_color = '#0b4f6c'  # dark teal for Desktop
fig, ax = plt.subplots(figsize=(12,5))
bar_height = 0.25  # Reduced bar height

# positions: place GG on top and Chat below with reduced spacing
y_pos = np.array([0.5, 0])  # Reduced spacing between bars

# stacked segments: first Mobile, then Desktop
mobile_vals = [gg_pct[0], chat_pct[0]]
desktop_vals = [gg_pct[1], chat_pct[1]]
ax.barh(y_pos, mobile_vals, height=bar_height, color=mobile_color, label='Mobile')
ax.barh(y_pos, desktop_vals, left=mobile_vals, height=bar_height, color=desktop_color, label='Desktop')

# annotate each segment: centered horizontally and vertically inside the segment
for i, yv in enumerate(y_pos):
    # Mobile segment
    m = mobile_vals[i]
    if m > 0:
        ax.text(m/2, yv, f'{m:.1f}%', va='center', ha='center', fontsize=12, color='black', weight='600')
    # Desktop segment
    d = desktop_vals[i]
    if d > 0:
        ax.text(m + d/2, yv, f'{d:.1f}%', va='center', ha='center', fontsize=12, color='white', weight='600')

# set y labels and limits
ax.set_yticks(y_pos)
ax.set_yticklabels(['GG Condition', 'Chat Condition'], fontsize=14)
ax.set_xlim(0, 100)
ax.set_xticks([0, 25, 50, 75, 100])
ax.set_xlabel('Percentage of Participants (%)')
ax.set_title('What kind of device participants used for the experiment?', fontsize=16, pad=50)

# remove top, right, left spines for a clean journalistic look
for spine in ['top', 'right', 'left']:
    ax.spines[spine].set_visible(False)

# legend centered between title and bars with larger markers and font
legend = ax.legend(loc='upper center', bbox_to_anchor=(0.5, 1.15), ncol=2, frameon=False, fontsize=12, markerscale=1.5)
for handle in legend.legend_handles:
    handle.set_height(12)
    handle.set_width(12)

plt.tight_layout()
plt.show()
```
