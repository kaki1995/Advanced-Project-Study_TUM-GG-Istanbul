```python
# Import relevant libaries

import pandas as pd
import os
from pathlib import Path
import openpyxl  # Ensure openpyxl is installed
import numpy as np
```

## 1. Match df_merged and df_after_standard datates

```python
# Load the df_merged.csv file
data_dir = Path('chat_exports')
df_merged = pd.read_csv(data_dir / 'df_merged.csv', sep=',')
print(f"df_merged shape: {df_merged.shape}")
# Preview the dataset
df_merged.head()
```

```python
# Load the df_after_standard.xlsx file
df_after_standard = pd.read_excel(data_dir / 'df_after_standard.xlsx')
print(f"df_after_standard shape: {df_after_standard.shape}")
# Preview the dataset
df_after_standard.head()
```

```python
# Concatenate selected columns from df_after_standard into df_merged at specified positions
def insert_after(df, col, new_cols):
    idx = df.columns.get_loc(col) + 1
    for i, new_col in enumerate(new_cols):
        df.insert(idx + i, new_col, df_after_standard[new_col])
    return df
df_concat = df_merged.copy()
df_concat = insert_after(df_concat, "q5_nationality", ["q5_nationality_sd"])
df_concat = insert_after(df_concat, "q6_residence", ["q6_residence_sd"])
df_concat = insert_after(df_concat, "q10_search_tools", ["q10_tool_1", "q10_tool_2", "q10_tool_3", "q10_tool_4"])
df_concat = insert_after(df_concat, "q12_brand_model", ["q12_brand_model_1_sd", "q12_brand_model_2_sd", "q12_brand_only"])
df_concat = insert_after(df_concat, "q13_storage", ["q13_storage_1_sd", "q13_storage_2_sd"])
df_concat = insert_after(df_concat, "q14_color", ["q14_color_1_sd", "q14_color_2_sd", "q14_color_3_sd", "q14_color_4_sd", "q14_color_5_sd"])
df_concat = insert_after(df_concat, "q15_price", ["q15_price_euro", "q15_price_range"])
print(f"df_concat shape: {df_concat.shape}")
df_concat.head()
```

```python
# Split q18_important_features into 3 columns and q39_contradictory_info into max columns, keeping originals
import ast
def split_list_column(df, col, prefix, n_cols):
    def parse_list(val):
        if isinstance(val, str):
            try:
                return ast.literal_eval(val)
            except:
                return [val]
        return val
    split_data = df[col].apply(parse_list).apply(lambda x: x if isinstance(x, list) else [x])
    for i in range(n_cols):
        df.insert(df.columns.get_loc(col)+1+i, f"{prefix}_{i+1}", split_data.apply(lambda x: x[i] if i < len(x) else None))
    return df
# For q18_important_features, always split into 3 columns
df_concat = split_list_column(df_concat, "q18_important_features", "q18_feature", 3)
# For q39_contradictory_info, find max items and split accordingly
max_items_q39 = df_concat["q39_contradictory_info"].apply(lambda x: len(ast.literal_eval(x)) if isinstance(x, str) and x.startswith('[') else 1).max()
df_concat = split_list_column(df_concat, "q39_contradictory_info", "q39_info", max_items_q39)
print(f"df_concat shape after splitting: {df_concat.shape}")
df_concat.head()
```
